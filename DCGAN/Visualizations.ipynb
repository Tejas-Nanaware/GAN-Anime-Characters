```python
import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
# MacOS matplotlib kernel issue
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
train_datagen = image.ImageDataGenerator(rescale=1./255)
train_gen = train_datagen.flow_from_directory('.', classes=['anime_face'], target_size = (128, 128), batch_size = 28, color_mode='rgb')
```

```python
num_images = train_gen[0][0].shape[0]
fig = plt.figure(figsize=(15,7))
columns = 7
rows = np.ceil(num_images / columns)
for i in range(num_images):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(train_gen[0][0][i])
plt.show(block=True)
```
